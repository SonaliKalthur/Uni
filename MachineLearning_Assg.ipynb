```python
#Calculate covariance matrix using raw formula(not using numpy cov)
print(strftime("%Y-%m-%d %H:%M:%S"))
mean_value = np.mean(X, axis=0)

#covariance = (x-mean(x).(y-mean(y))/N-1 for sample set(covariance))

cov_raw = (X - mean_value).T.dot((X - mean_value)) / (X.shape[0]-1)
print('Covariance matrix not using numpy cov \n%s' %cov_raw)
print(strftime("%Y-%m-%d %H:%M:%S"))
```

```python
#Calculate covariance matrix using numpy cov function just to check 
print(strftime("%Y-%m-%d %H:%M:%S"))
covariance_numpy = np.cov(X.T) 
print("Covariance Value\n",covariance_numpy)
print(strftime("%Y-%m-%d %H:%M:%S"))
```

# Eigen Decomposition

```python
#Eigen Decomposition

print("Start Eigen Decomposition: ",strftime("%Y-%m-%d %H:%M:%S"))

#To calculate eigen values and eigen vector using linalg function
eigen_value , eigen_vector = np.linalg.eig(covariance_numpy)
print("Eigen Values\n",eigen_value)
print("Eigen Vector\n",eigen_vector)

#Sort the Eigen values from descending order such that to find the highest eigen value
print("Start time for Eigen sort function: ",strftime("%Y-%m-%d %H:%M:%S"))
eigen_tuple = []

for i in range(len(eigen_value)):
    eigen_tuple.append((eigen_value[i],eigen_vector[:,i]))
    
#sort the eigen values from descending order
eigen_tuple.sort()
eigen_tuple.reverse()
print("Sort the Eigen Values and print in reverse order")
for i in eigen_tuple:
    print(i[0])
print("End time of Eigen sort function: ",strftime("%Y-%m-%d %H:%M:%S"))
```

```python
print(strftime("%Y-%m-%d %H:%M:%S"))

sum_eigen = sum(eigen_value)
var_eigen = [(i/sum_eigen)*100 for i in sorted(eigen_value,reverse=True)]
cum_var_eigen = np.cumsum(var_eigen)
print(cum_var_eigen[1000],cum_var_eigen[1997])
```

# SVD

```python
print(strftime("%Y-%m-%d %H:%M:%S"))
u,s,v = np.linalg.svd(X.T)
e2=s**2/len((s)-1)
sum_svd = sum(e2)
var_svd = [(i/sum_svd)*100 for i in sorted(e2,reverse=True)]
cum_var_svd = np.cumsum(var_svd)
print(cum_var_svd[1000],cum_var_svd[1997])
print(strftime("%Y-%m-%d %H:%M:%S"))

```
